## Introduccin

En estadstica, inteligencia artificial, aprendizaje de mquina y ciencias en general, siempre ha sido una labor importante encontrar una funcin $f$ que 'ajuste' a un conjunto de datos.

Supongamos que como resultado de un experimento se obtienen 4 puntos $(x, y) : (1, 6), (2, 5), (3, 7), (4, 10)$. Y se busca encontrar una recta de la forma

$$y=\alpha+\beta x \qquad \text{(ecuacin explcita de la recta)} $$

Que se ajuste lo mejor posible a este conjunto de datos, la idea es encontrar la recta que pase lo ms cercano posible a todos estos puntos. Para encontrar dicha recta vamos a tratar de **minimizar el error que existe entre los datos reales y la aproximacin** dada por la recta.

Si lo pensamos en forma de sistema de ecuaciones, se buscan valores $\alpha$, $\beta$ **que se aproximen a solucionar el sistema lineal** (sobredeterminado).

$$sistema = \begin{cases} 
\alpha+1\beta&=&6\\
\alpha+2\beta&=&5\\
\alpha+3\beta&=&7\\
\alpha+4\beta&=&10
\end{cases}$$

## 2. Mnimos Cuadrados

La forma de solucionar el problema antes mencionado, se conoce como 'ajuste por mnimos cuadrados'. Y la idea de este algoritmo es tratar de minimizar la suma de los cuadrados de los errores.



El error en cada punto entre la curva aproximada y los datos, es la suma del lado izquierdo y derecho de las ecuaciones del sistema $Ax=b$, es decir

$F\left(\alpha,\beta\right)	=	\left[6-\left(\alpha+1\beta\right)\right]^{2}+\left[5-\left(\alpha+2\beta\right)\right]^{2}+\left[7-\left(\alpha+3\beta\right)\right]^{2}+\left[10-\left(\alpha+4\beta\right)\right]^{2}
	=	4\alpha^{2}+30\beta^{2}-56\alpha-154\beta+20\alpha\beta+210$
    

El mnimo de la funcin F se determina mediante las derivadas parciales de (tambin conocido como el gradiente) $F\left(\alpha,\beta\right)$ igualado a cero

$$\frac{\delta F}{\delta\alpha}=0=8\alpha+20\beta-56\quad y\quad\frac{\delta F}{\delta\beta}=0=20\alpha+60\beta-154 \tag{1})$$

El sistema (1) es conocido como las **ecuaciones normales** (del sistema original $Ax=b$) debido a que se emplean las derivadas parciales igualadas a cero Que hay de diferente en este nuevo sistema $A'x=b$?.

La solucin de este sistema es $\alpha=3.5$ y $\beta=1.4$, de tal manera que la recta que mejor aproxima linealmente todos los datos de este ejemplo es $$y=3.5+1.4x$$


De manera general el problema antes descrito lo podemos expresar de la siguiente forma. Sea $A\in\mathbb{R}_{m\times n}$ y el rango de $A=n$ ( rango(A) renglones linealmente independientes),  $\vec{x}$,  $\vec{b}\in\mathbb{R}^{n}$.

Se tiene un sistema $A\vec{x}=\vec{b}$ que no tiene solucin (ya que es sobredeterminado), lo que se busca es encontrar los coeficientes $x_{i}$ que hagan que la ecuacin buscada se ajuste lo mejor posible en trminos de minimizacin cuadrtica.

En otras palabras, se trata de solucionar $A\vec{x}\approx\vec{b}$, por lo que se busca un vector $\hat{x}$ que minimice la funcin $F\left(\vec{x}\right)$, es decir

$$\hat{x}=min(F(\vec{x}))$$

Como $F(\vec{x})=\left\Vert A\vec{x}-\vec{b}\right\Vert _{2}^{2}$ y dado que se busca minimizar $F\left(\vec{x}\right)$ se emplea el gradiente ($\nabla$ operador nabla) igualado a cero.

Este problema de minimizacin tiene solucin nica ya que el rango de $A$ es $n$, finalmente el nuevo sistema es conocido como las ecuaciones normales que caracterizan al sistema original. 


## 3. Ecuaciones Normales.

Una forma de expresar (y mostrar el desarrollo de como llegar) a las ecuaciones normales de un sistema $A\vec{x}=\vec{b}$ es empleando el gradiente.

Sin embargo existe una forma mas sencilla y compacta de expresar el sistema de ecuaciones normales que caracteriza a dicho sistema.

Sea $A\in\mathbb{R}_{m\times n}$ (con $m\geq n$) y $rango(A)=n$, $b\in\mathbb{R}^{m}$ y $\vec{x}$ la solucin que minimiza el residual $r=\left\Vert A\vec{x}-\vec{b}\right\Vert _{2}^{2}$, entonces las ecuaciones normales estn dadas por

$$A^{T}A\vec{x}=A^{T}\vec{b}$$


La idea principal para resolver problemas de minimizacin recae en el concepto de proyeccin ortogonal. Suponiendo que el vector $\vec{b}$ no pertenece al subespacio generado por las columnas de la matriz A (o $Im(A)$), denotamos por 

$$P_{A}:R^{m}\longrightarrow Im(A)$$

A la proyeccin ortogonal que mapea $R^{m}$ sobre $Im(A)$. Entonces el valor $\vec{x}$ que minimiza el residual $\vec{r}=A\vec{x}-\vec{b}$ es aquel que satisface $A\vec{x}=P_{A}\vec{b}$.


En otras palabras el residual $\vec{r}=A\vec{x}-\vec{b}$ debe ser ortogonal al subespacio generado por las columnas de A.

![subespacio generado](https://docs.google.com/uc?export=download&id=1rHJ4JiJEVE8kFZGwcl3efzVZf5mnhYwq)


**TEOREMA** 

Sea $A\in\mathbb{R}_{m\times n}$ (con $m\geq n$) y $rango(A)=n$, $b\in\mathbb{R}^{m}$. El residual $\vec{r}=A\vec{x}-\vec{b}$ es ortogonal a $Im(A)$ (subespacio generado por los vectores columna de la matriz) A $\hspace{2mm}$ si y solo si $\hspace{2mm}$ $A^{T}A\vec{x}=A^{T}\vec{b}$

**Demostracin.** 

El residual $\vec{r}$ es ortogonal a $Im(A)$ si y solo si $\vec{a}_{i}^{T}\vec{r}=0$ $\forall i=1,\ldots,n$. Por lo tanto $\vec{r}$ es ortogonal a $Im(A)$ si y solo si

$$A^{T}\left(A\vec{x}-\vec{b}\right)=\vec{0} \tag{2} $$

Pero de (2) tenemos que

$$A^{T}\left(A\vec{x}-\vec{b}\right)=A^{T}A\vec{x}-A^{T}\vec{b}=0\Longrightarrow A^{T}A\vec{x}=A^{T}\vec{b}$$

Por lo tanto $\vec{r}$ es ortogonal a $Im(A)$ si y solo si

$$A^{T}A\vec{x}=A^{T}\vec{b}$$

Encuentre las ecuaciones normales del sistema $A\vec{x}=\vec{b}$ y encuentre la ecuacin de la recta que se ajusta de mejor manera a los puntos: $\left(x,y\right)$: $ \color{red}{(1,6),(2,5),(3,7),(4,10)}$ 

$\color{blue}{Solucin}$: Dados los puntos (x,y) de este ejemplo, se busca minimizar la funcin que pase por todos ellos entonces el sistema $A\vec{x}=\vec{b}$ es el siguiente

$$A\vec{x}=\left(\begin{array}{cc}
1 & 1\\
1 & 2\\
1 & 3\\
1 & 4
\end{array}\right)\left(\begin{array}{c}
\alpha\\
\beta
\end{array}\right)=\left(\begin{array}{c}
6\\
5\\
7\\
10
\end{array}\right) $$

As que empleando el teorema antes mencionado, las ecuaciones normales de este sistema son $A^{T}A\vec{x}=A^{T}\vec{b}$

$$\left[\left(\begin{array}{cccc}
1 & 1 & 1 & 1\\
1 & 2 & 3 & 4
\end{array}\right)\left(\begin{array}{cc}
1 & 1\\
1 & 2\\
1 & 3\\
1 & 4
\end{array}\right)\right]\left(\begin{array}{c}
\hat{x}_{1}\\
\hat{x}_{2}
\end{array}\right)=\left(\begin{array}{cccc}
1 & 1 & 1 & 1\\
1 & 2 & 3 & 4
\end{array}\right)\left(\begin{array}{c}
6\\
5\\
7\\
10
\end{array}\right) $$

Al resolver este nuevo sistema $(A\hat{x}=\vec{b})$ se obtiene que el vector solucin es $\left(3.5,1.4\right)$, as que la recta que mejor se ajusta a los datos de este ejemplo es

$$ y=3.5+1.4x$$

## 4. Factorizacin QR.

Como ya se ha visto a lo largo de este curso, es recurrente tratar de resolver un sistema del tipo $A\vec{x}=\vec{b}$ y dado que el clculo de la matriz inversa de $A$ es un proceso costoso, se buscan alternativas.

La matriz asociada al sistema de ecuaciones normales $\left(A^{T}A\right)$ vista en la seccin anterior, tiene peculiares caractersticas ($\color{blue}{Simtrica y Positiva Definida}$) que la convierte en candidato a diferentes tipos de factorizaciones, en este caso la factorizacin QR.

**HECHO**

Una matriz $A\in\mathbb{R}_{n\times n}$

1. Es simtrica si: $A=A^{T}$

2. Es positiva definida si: Todos sus determinantes superiores izquierdos de $i\times i,\quad i=1,\ldots,n$ son positivos. 


**Teorema**

Sea $A\in\mathbb{R}_{m\times n$} (con $m\geq n$) y $rango\left(A\right)=n$ existe una nica factorizacin QR tal que

$$A=QR\quad con \quad r_{ii}>0\quad i=1,\ldots,n$$

### Mtodo de Gram-Schmidt.

La idea detrs de la factorizacin QR es construir una base ortonormal $\left(q_{1},q_{2},\ldots,q_{n}\in R^{m}\right)$ para el subespacio generado por los vectores columna de A.

*  Se comienza utilizando el primer vector columna de $A\quad(a_{1})$. Es decir $q_{1}=\frac{a_{1}}{\left\Vert a_{1}\right\Vert _{2}}$.


* A continuacin se substrae del vector $a_{2}$ su coeficiente en la direccin de $q_{1}$, es decir $\left(q_{1}^{T}\cdot a_{1}\right)\cdot q_{1}$. Dando como resultado un vector $q_{2}$, que es ortogonal a $q_{1}$. Finalmente se normaliza el vector $q_{2}$ para volverlo ortonormal.


* El proceso se repite con el resto de los vectores columna de $A=\left[a_{1},a_{2},\ldots,a_{n}\right]$


En trminos matemticos, el proceso se vera de la siguiente manera.

### Algoritmo.

1. $q_{1}=\frac{v_{1}}{\left\Vert v_{1}\right\Vert _{2}}\quad con\quad v_{1}=a_{1}$

2. $ q_{2}=\frac{v_{2}}{\left\Vert v_{2}\right\Vert _{2}}\quad con\quad v_{2}=a_{2}-\left(q_{1}^{T}\cdot a_{2}\right)\cdot q_{1} $

3. $q_{3}=\frac{v_{3}}{\left\Vert v_{3}\right\Vert _{2}}\quad con\quad v_{3}=a_{3}-\left(q_{1}^{T}\cdot a_{3}\right)\cdot q_{1}-\left(q_{2}^{T}\cdot a_{3}\right)\cdot q_{2} $


4. En general el j-esimo paso, suponiendo conocidos $q_{1},\ldots,q_{j-1}$, es un vector $q_{j}$ ortonormal a todos los vectores ya conocidos, es decir

$$ q_{j}=\frac{v_{j}}{\left\Vert v_{j}\right\Vert _{2}}\quad con\quad v_{j}=a_{j}-\left(q_{1}^{T}\cdot a_{j}\right)\cdot q_{1}-\cdots-\left(q_{j-1}^{T}\cdot a_{j}\right)\cdot q_{j-1}$$


5. En forma de suma, el vector $v_{j}$ se ve de la siguiente manera

$$v_{j}=a_{j}-\sum_{i=1}^{j-1}\left(q_{i}^{T}\cdot a_{j}\right)\cdot q_{i}$$

### Propiedades de QR 

La factorizacin $QR$ implica que la matriz $Q$ es una matriz ortonormal y la matriz $R$ es una matriz triangular superior. Estos factores tiene muchas ventajas.

**Hecho.** 

Sea $A\in\mathbb{R}_{n\times n}$

1. Si $Q$ es ortogonal entonces $Q^{T}Q=Id$

2. $QR=A\Longrightarrow R=Q^{T}A.$

3. Dado que de las ecuaciones normales definidas por $A^{T}A\vec{x}=A^{T}\vec{b}$, garantizan que $A^{T}A$ es una matriz positiva definida, entonces $A^{T}A$ se pude factorizar en $QR$.

4. Sea $A'=A^{T}A$ y $b'=A^{T}\vec{b}$. Entonces

$$A'\vec{x}=b'\Longrightarrow QR\vec{x}=b'\Longrightarrow Q^{T}QR\vec{x}=Q^{T}b' \Longrightarrow R\vec{x}=Q^{T}b'$$

Y resulta que es un sistema triangular que puede ser resuelto fcilmente mediante su buen amigo, substitucin hacia atrs.

### Procedimiento general para Mnimos Cuadrados.

Ya que se entendido el por que es til entender el proceso de mnimos cuadrados y como es que las factorizaciones ayudan a resolverlo, ahora lo que resta es automatizar todo el proceso que lleva a resolver el problema general, es decir.

* Se tiene una lista de puntos de la forma $\left(x,y\right)$, donde cada punto representa datos reales de algn fenmeno (estadstica, gentica, geofsica, fsica biomdica, etc).


* A continuacin se encuentran las ecuaciones normales del sistema empleando la igualdad $A^{T}A\vec{x}=A^{T}\vec{b}=b'.$


* Aplicamos alguna factorizacin, por ejemplo Cholesky $A^{T}A=A\text{'}=L^{T}L$  $A\text{'}=QR$ empleando el proceso de Gram-Schmidt para encontrar una base ortonormal para el subespacio generado por los vectores columna de A'.


* Aplicando las propiedades de $L^{T}$ y $L$  $Q$ y $R$ se resuelve el nuevo sistema (mediante substitucin hacia atrs), es decir $QR\vec{x}=b'$ $\Longrightarrow$  $R\vec{x}=Q^{T}b'$.


* Una vez encontrados los coeficientes $\alpha$ y $\beta$ de la recta que minimizan los errores cuadrticos, ahora es posible graficar dicha recta e incluso interpolar.
