### Introducci√≥n

Dentro de este notebook, aprenderemos qu√© es un vector, as√≠ algunas operaciones con estos.

El t√©rmino vector proviene del lat√≠n vector, vectoris, cuyo significado es 'el que conduce', o 'el que transporta'.
Se le denomina vector de dimensi√≥n $n$ a una tupla de n√∫meros reales (componentes del vector). El conjunto de todos los vectores de dimensi√≥n $n$, se representa como $R^n$. Es decir, un vector $\vec{v}$ perteneciente a un espacio vectorial $R^n$ se representa de la siguiente manera:\
$\vec{v}=(a_{1},a_{2},a_{3},...,a_{n})$, donde $\vec{v} \in R^n$

   + **¬øQu√© es un vector?**\
Dentro del √°rea de la f√≠sica, se le conoce como vector a un segmento de recta en el espacio que parte de un punto hacia otro, es decir, que tiene direcci√≥n y sentido. Los vectores en f√≠sica tienen por funci√≥n expresar las llamadas magnitudes vectoriales. En computaci√≥n, un vector representa una lista de n√∫meros que pueden tomar muchos valores.
En matem√°ticas se define vector como un elemento de un espacio vectorial. Esta noci√≥n es m√°s abstracta y para muchos espacios vectoriales no es posible representar sus vectores mediante el m√≥dulo y la direcci√≥n, en particular los espacios de dimensi√≥n infinita sin producto escalar no son representables de ese modo. Los vectores en un espacio eucl√≠deo se pueden representar geom√©tricamente como segmentos de recta $R$, en el plano $R^{2}$ (bidimensional), o en el espacio $R^{3}$ (tridimensional).

Mientras en el √°rea de la computaci√≥n, los vectores son una forma de almacenar datos que permiten contener una serie de valores del mismo tipo, cada uno de los valores contenidos tiene una posici√≥n asociada que se usar√° para accederlos. Est√° posici√≥n o √≠ndice ser√° siempre un n√∫mero entero positivo.


   + **Suma y resta de vectores**\
\
Para sumar dos vectores $\vec{a}$  y $\vec{b}$ se suma $\vec{a}$ con el vector $\vec{b}$, es decir, se suman los componentes de cada vector:
\begin{equation}
\vec{a}+\vec{b} = (a_{x}+b_{x},a_{y}+b_{y},a_{z}+b_{z})
\end{equation}
Por ejemplo:\
Sean $\vec{a}=(3,2,-4)$ y $\vec{b}=(-3,2,7)$, calcula el vector $\vec{a}+\vec{b}$
\begin{equation}
\vec{a}+\vec{b} = (3+(-3),2+2,-4+7)=(0,4,3)
\end{equation}
\
\
Para restar dos vectores $\vec a$ y $\vec b$ se suma $\vec a$ con el opuesto del vector $\vec b$, es decir:}
\begin{equation}
\vec a- \vec b = (\vec a_{x}- \vec b_{x}, \vec a_{y}- \vec b_{y},\vec a_{z}- \vec b_{z})
\end{equation}
Por ejemplo:\
Sea $\vec a=(5,2,4)$ y $\vec b=(-3,5,9)$, calcular el vector $\vec a- \vec b$
\begin{equation}
\vec a- \vec b=(5-(-3),2-5,4-9)=(8,-3,-5)
\end{equation}

Sumando y Restando Vectores, al vector $\vec x$ se le desea sumar otro vector $\vec y$

```
#Declaramos nuestro vector y
y = np.array([2, -1,])
print('Nuestro vector x es:\n', x)
print('Nuestro vector y es:\n', y, '\n')
#Utilizando los operadores aritmeticos + y -, realizamos las operaciones necesarias
print(f'La suma de nuestros vectores X y Y es: {x+y}')
print(f'La resta de nuestros vectores X y Y es: {x-y}')
```

 + **Multiplicaci√≥n de un Vector por un escalar**\
\
Al multiplicar un vector $\vec{x}$ por un escalar $\lambda$, obtenemos como un nuevo vector $\vec{y}=(\lambda)(\vec{x})$ que cumple con las siguientes caracter√≠sticas:\
Si $\lambda$ es:\
\
~ Positivo $\vec{x}$ y $\vec{y}$ tendr√°n el mismo sentido.\
~ Negativo $\vec{x}$ y $\vec{y}$ tendr√°n diferente sentido.\
\
Por ejemplo:\
Sea $\vec{x} =(2,4,1,2)$ y $\lambda=2$\
\begin{equation}
(\vec{x})(\lambda) = 2(2,4,1,2) = (4,8,2,4)
\end{equation}

No importa cuantas componentes tenga nuestro vector, el escalar $\lambda$ multiplica cada una de las componentes del vector. Un ejemplo en 2 dimensiones es el siguiente.

Para multiplicar un vector $\vec{x}$ por un escalar $\lambda$, √∫nicamente definimos dicho escalar y realizamos el producto con nuestro arreglo.

```
#Asignamos una variable k representando nuestro escalar
k = 2
#Declaramos nuestro vector
vector = np.array([1, 3])
#Realizamos la multiplicaci√≥n
producto = vector*k
print('Vector: ', vector, '\n')
print('Escalar: ', k, '\n')
print(f'El producto es: {producto}')
```

Cuando el escalar es positivo

```
par = [{'title':'Producto (+Escalar)(Vector)','ylim':(-3,9), 'xlim':(-0.5,3.5)}]
graf = vis.planoCartesiano(par=par)
graf.plot(x=np.array([0, 1]), y=[0,3], par={'label':'Vector', 'ls':'dashdot','lw':1, 'color':'blue'})
graf.plot(x=np.array([0, 2]), y= [0,6], par={'label':'Producto', 'ls':':','lw':2, 'color':'red'})
graf.scatter(x=np.array([0,1]), y=[0,3], par={ 'ls':'dashdot','lw':4, 'color':'blue','marker':'v'})
graf.scatter(x=np.array([0, 2]), y=[0,6], par={ 'ls':':','lw':6, 'color':'red','marker':'v'})
graf.legend()
graf.show()
```

Cuando el escalar es negativo, se tienen direcciones distintas

```
k = -2
producto = vector*k
print('Escalar: ',k)
print(f'El producto es: {producto}')
```

```
par = [{'title':'Producto (-Escalar)(Vector)','ylim':(-7,7), 'xlim':(-2,2)}]
graf = vis.planoCartesiano(par=par)
graf.plot(x=np.array([0, 1]), y=[0,3], par={'label':'Vector', 'ls':'dashdot','lw':1, 'color':'blue'})
graf.plot(x=np.array([0, -2]), y= [0,-6], par={'label':'Producto', 'ls':':','lw':2, 'color':'red'})
graf.scatter(x=np.array([0,1]), y=[0,3], par={ 'ls':'dashdot','lw':4, 'color':'blue','marker':'v'})
graf.scatter(x=np.array([0, -2]), y=[0,-6], par={ 'ls':':','lw':6, 'color':'red','marker':'<'})
graf.legend()
graf.show()
```

### Vector Unitario
Un vector unitario hace referencia al vector cuyo m√≥dulo (distancia con respecto al origen) es igual a 1. Cabe recordar que el m√≥dulo es la cifra coincidente con la longitud cuando el vector se representa en un gr√°fico. El m√≥dulo, de este modo, es una norma de la matem√°tica que se aplica al vector que aparece en un espacio eucl√≠deo.
Para hallar un vector unitario a partir de cualquier vector, hay que dividir este √∫ltimo por su m√≥dulo.
### Ejercicio:
Si $\vec{v}$ es un vector de componentes (3,4), encontrar un vector unitario de su misma direcci√≥n y sentido, dada la f√≥rmula:
\begin{equation}
\vec{u} = \frac{1}{|\vec{v}|} \vec{v}
\end{equation}
- Calcular magnitud de $\vec{v}$
\begin{equation}
|\vec{u}| = \sqrt{(3)^2 + (4)^2} = \sqrt{25} = 5
\end{equation}
Sustituyendo:
\begin{equation}
\vec{u} = \frac{1}{5}(3,4) = (\frac{3}{5},\frac{4}{5})
\end{equation}
Visto desde c√≥digo:

```
vector_unitario = []
vector = [3,4]
#Calculando magnitud
magnitud = math.sqrt(vector[0]**2 + vector[1]**2)
#En el siguiente ciclo, almacenamos los valores de nuestro vector
for i in range(len(vector)):
  vector_unitario.append(vector[i]*(1/magnitud))
print('Vector original', vector)
print('Nuestro vector unitario es: ',vector_unitario)
```

```
par = [{'title':'Vector Unitario','ylim':(0,4.5), 'xlim':(0,3.5)}]
graf = vis.planoCartesiano(par=par)
graf.plot(x=np.array([0, 3]), y=[0,4], par={'label':'Vector', 'ls':'dashdot','lw':1, 'color':'blue'})
graf.plot(x=np.array([0, 0.6]), y= [0,0.8], par={'label':'Vector Unitario', 'ls':':','lw':2, 'color':'red'})
graf.scatter(x=np.array([0, 3]), y=[0,4], par={ 'ls':'dashdot','lw':4, 'color':'blue','marker':'v'})
graf.scatter(x=np.array([0, 0.6]), y=[0,0.8], par={ 'ls':':','lw':6, 'color':'red','marker':'v'})
graf.legend()
graf.show()
```

```
par = [{'title':'Problema de Desplazamiento','ylim':(41,-11), 'xlim':(-0.25,16)}]
vectores = [[15, 40], [15, -10]]
graf = vis.planoCartesiano(par=par)
graf.plot(x=np.array([0, 15]), y=np.array([0, 40]), par={'label':'', 'ls':'dashdot','lw':1, 'color':'blue'})
graf.plot(x=np.array([0, 15]), y= np.array([0, -10]), par={'label':'', 'ls':':','lw':2, 'color':'red'})
graf.scatter(x=np.array([0, 15]), y=np.array([0, 40]), par={ 'ls':'dashdot','lw':3, 'color':'blue'})
graf.scatter(x=np.array([0, 15]), y= np.array([0, -10]), par={ 'ls':':','lw':5, 'color':'red'})
graf.show()

```

$$|d| = \sqrt{225 + 1600}$$
\
$$|d| = \sqrt{1825}$$

$$Desplazamiento\ totla\ del\ bal√≥n\ =\ |d| = 42.72\ yardas$$
